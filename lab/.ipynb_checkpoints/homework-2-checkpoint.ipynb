```python
import random

import numpy as np
import torch
import pandas as pd

from lab.tutorial_2b.vfl import BottomModel, VFLNetwork
import io
import sys
import re

seed = 0
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed(seed)
torch.mps.manual_seed(seed)
```

```python
from sklearn.preprocessing import MinMaxScaler

# loading and preprocessing 

df = pd.read_csv("../lab/tutorial_2a/heart.csv")
categorical_columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
numerical_columns = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

df[numerical_columns] = MinMaxScaler().fit_transform(df[numerical_columns])  # scale numerical features/data
encoded_df = pd.get_dummies(df, columns=categorical_columns)  # one hot encode categorical features/data

all_features = list(df.columns)
all_features.remove("target")

X = encoded_df.drop("target", axis=1)  # input features
Y = pd.get_dummies(encoded_df[['target']], columns=['target'])  # Y (outputs) are one-hot labels
```

```python
permutations = []

# permutations generated
for i in range(3):
    np.random.seed(seed + i)
    permutation = np.random.permutation(all_features)
    permutations.append(list(permutation))

# print(len(permutations[0]))
print(permutations[0][:3])
print(permutations[1][:3])
print(permutations[2][:3])
# print(permutations)
```

```python
def train_vfl_with_permutation(X, Y, features, categorical_columns, clients=4,
                               epochs=300, batch_size=64, train_test_threshold=0.8):
    """
    Trains the VFL model on a single permutation of features.
    Returns the loss curve and final test accuracy.
    """
    X_train, X_test = X.loc[:int(train_test_threshold * len(X))], X.loc[int(train_test_threshold * len(X)) + 1:]
    Y_train, Y_test = Y.loc[:int(train_test_threshold * len(Y))], Y.loc[int(train_test_threshold * len(Y)) + 1:]

    # compute features per client
    features_per_client = (clients - 1) * [len(features) // clients]
    features_per_client.append(len(features) - sum(features_per_client))

    # assign features to each client
    client_feature_names = []
    start = 0
    for num_feats in features_per_client:
        client_feature_names.append(features[start:start + num_feats])
        start += num_feats

    # map to one-hot encoded features
    encoded_feature_names = list(X.columns)
    updated_client_features = []
    for client_feats in client_feature_names:
        updated = []
        for col in client_feats:
            if col not in categorical_columns:
                updated.append(col)
            else:
                for name in encoded_feature_names:
                    if name.startswith(col + "_"):
                        updated.append(name)
        updated_client_features.append(updated)

    # model architecture
    outs_per_client = 2
    bottom_models = [BottomModel(len(in_feats), outs_per_client * len(in_feats)) for in_feats in
                     updated_client_features]
    Network = VFLNetwork(bottom_models, n_outs=2)

    # capture training logs
    old_stdout = sys.stdout
    sys.stdout = mystdout = io.StringIO()

    # train
    Network.train_with_settings(epochs, batch_size, clients, updated_client_features, X_train, Y_train)

    sys.stdout = old_stdout
    output = mystdout.getvalue()

    # parse loss
    losses = []
    for line in output.split("\n"):
        match = re.search(r"Epoch: \d+ Train accuracy: ([\d.]+)% Loss: ([\d.]+)", line)
        if match:
            losses.append(float(match.group(2)))

    avg_loss = sum(losses) / len(losses)

    # test
    acc, _ = Network.test(X_test, Y_test)

    print(f"Final Train Loss: {losses[-1]:.4f}")
    print(f"Average Train Loss: {avg_loss:.4f}")
    print(f"Test Accuracy: {acc * 100:.2f}%")

    return losses, acc.item()
```

```python
losses_all = []
accuracies_all = []

for perm_idx, permuted_features in enumerate(permutations):
    print(f"\nTraining on permutation {perm_idx + 1}...")

    losses, accuracies = train_vfl_with_permutation(
        X=X,
        Y=Y,
        features=permuted_features,
        categorical_columns=categorical_columns,
        clients=6
    )

    losses_all.append(losses)
    accuracies_all.append(accuracies)
```

```python
import matplotlib.pyplot as plt

step = 10  # plot every 10th point

plt.figure(figsize=(10, 5))

for i, losses in enumerate(losses_all):
    perm_idx = i % 3  # 0, 1, 2

    downsampled_losses = losses[::step]
    downsampled_epochs = list(range(0, len(losses), step))

    plt.plot(downsampled_epochs, downsampled_losses, label=f'Perm {perm_idx + 1}')

plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.title("Training Loss per Epoch for Each Permutation")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
features_per_client = [3, 3, 3, 4]

for i, perm in enumerate(permutations):
    print(f"\nPermutation {i + 1}:")

    start = 0
    for client_idx, num_feats in enumerate(features_per_client):
        assigned_feats = perm[start:start + num_feats]
        print(f"  Client {client_idx} -> {assigned_feats}")
        start += num_feats
```

### 1.1 
##### Describe what impact (if any) the permutations have on the loss and explain the observed pattern.

The three different permutations produced distinct training loss curves, even though they followed a generally similar convergence pattern. All permutations gave a reduced training loss over time, but the initial convergence speed, smoothness, and final loss values varied slightly across them.

In the observed results, Permutation 1 demonstrates the fastest and smoothest convergence, followed by Permutation 2, whereas Permutation 3 shows a slower and more fluctuating training curve.

These differences are likely due to how informative features were grouped. In Vertical Federated Learning, each client trains a local (bottom) model on its own set of features. Then these intermediate outputs are passed to the shared top model for prediction. When more informative features are concentrated in a single client's local model, it can produce richer representations, which enhance the global model's ability to converge efficiently. In this experiment, some permutations may have grouped more informative features within the same client, enabling that clients local model to learn stronger representations (richer and more useful signal), which contributed to faster convergence and lower final loss.

To conclude, the way features are partitioned among clients plays a critical role in VFL. It directly affects the local learning dynamics, which also impacts the performance of the global model.

###  2.1
#### Explain your scheme for partitioning the feature space under increasing number of clients.

I used a simple and balanced scheme for partitioning the feature space under increasing number of clients. The function used ensures that each client receives approximately the same number of features, taking into account the total feature count. The function first calculates the base number of features each client should have, and then the remaining features that can not be evenly divided are distributed one by one to the first few clients. This guarantees that the difference in the number of features between any two clients is at most one. Features are assigned in the original order without any shuffling. This allows all features to be used without duplication or omission, while also avoiding having a client with too few features, for example only 1, which could significantly limit its contribution to the overall model.

### 2.2
#### Implement the above explained partition mechanism for the requested client counts for the discriminative VFL model.

```python
def split_features_evenly(features, n_clients):
    base = len(features) // n_clients
    remainder = len(features) % n_clients
    sizes = [base + 1 if i < remainder else base for i in range(n_clients)]

    splits = []
    start = 0
    for size in sizes:
        splits.append(features[start:start + size])
        start += size
    return splits


client_cases = [2, 4, 6, 8, 10]  # sanity check
for num_client in (client_cases):
    splits = split_features_evenly(all_features, num_client)
    split_s = []
    for split in splits:
        split_s.append(len(split))

    print(f"split structure for {num_client} clients: {split_s}")
```

```python
def train_vfl_with_clients(X, Y, features, categorical_columns, clients=4,
                           epochs=300, batch_size=64, train_test_threshold=0.8):
    """
    Trains the VFL model on a single permutation of features.
    Returns the loss curve and final test accuracy.
    """
    X_train, X_test = X.loc[:int(train_test_threshold * len(X))], X.loc[int(train_test_threshold * len(X)) + 1:]
    Y_train, Y_test = Y.loc[:int(train_test_threshold * len(Y))], Y.loc[int(train_test_threshold * len(Y)) + 1:]

    client_feature_names = split_features_evenly(features, clients)

    # map to one-hot encoded features
    encoded_feature_names = list(X.columns)
    updated_client_features = []
    for client_feats in client_feature_names:
        updated = []
        for col in client_feats:
            if col not in categorical_columns:
                updated.append(col)
            else:
                for name in encoded_feature_names:
                    if name.startswith(col + "_"):
                        updated.append(name)
        updated_client_features.append(updated)

    # model architecture
    outs_per_client = 2
    bottom_models = [BottomModel(len(in_feats), outs_per_client * len(in_feats)) for in_feats in
                     updated_client_features]
    Network = VFLNetwork(bottom_models, n_outs=2)

    # capture training logs
    old_stdout = sys.stdout
    sys.stdout = mystdout = io.StringIO()

    # train
    Network.train_with_settings(epochs, batch_size, clients, updated_client_features, X_train, Y_train)

    sys.stdout = old_stdout
    output = mystdout.getvalue()

    # parse loss
    losses = []
    for line in output.split("\n"):
        match = re.search(r"Epoch: \d+ Train accuracy: ([\d.]+)% Loss: ([\d.]+)", line)
        if match:
            losses.append(float(match.group(2)))

    avg_loss = sum(losses) / len(losses)

    # test
    acc, _ = Network.test(X_test, Y_test)

    print(f"Final Train Loss: {losses[-1]:.4f}")
    print(f"Average Train Loss: {avg_loss:.4f}")
    print(f"Test Accuracy: {acc * 100:.2f}%")

    return losses, acc.item()
```

```python
losses_all_clients = []
accuracies_all_clients = []

client_sizes = [2, 3, 4, 5, 6, 7, 8, 9, 10]
for idx, client_num in enumerate(client_sizes):
    print(f"\nTraining for {client_num} clients...")

    losses, accuracies = train_vfl_with_clients(
        X=X,
        Y=Y,
        features=all_features,
        categorical_columns=categorical_columns,
        clients=client_num
    )

    losses_all_clients.append(losses)
    accuracies_all_clients.append(accuracies)
```

### 2.4
#### State what impact (if any) adding more clients has on the loss and explain the observed pattern.

As we increase the number of clients in VFL, we preserve privacy by distributing features across clients. However, this comes at a cost to learning efficiency. Splitting the feature space more means that features which may hold strong correlations (e.g. "cholesterol" and "age") might end up on different clients. This fragmentation limits the ability of any individual local (bottom) model to capture richer, correlated feature representations.

As a result, the active party (which aggregates local outputs) receives weaker or less informative embeddings from each client, leading to slower convergence and higher overall loss. This is clearly observable in the training loss curves: the models trained with fewer clients (e.g. 2 or 3) converge faster and achieve lower final losses compared to models trained with more clients (e.g. 9 or 10), where loss curves flatten out earlier and remain higher. This suggests a trade-off between privacy granularity and predictive performance in VFL setups.

### Additional reflection
If there are clients with only 1 feature (when number of clients is > 7 in our case, and we have a small feature space), maybe we should assign them one more random feature, so they have at least something more correlated.
Considering that our feature space is quite small, if we have a number of clients higher than 7, then we end up having clients with only 1 feature, which may produce weak embeddings, hurt convergence and learn useless representations. So, I would like to experiment with this case and see what happens if these single-feature clients get assigned 1 more feature, so they have at least 2. 


```python
import random

def split_features_with_minimum(features, n_clients, min_features=2, seed=42):
    """
    Splits features across clients while ensuring each client gets at least `min_features`.
    If the number of features is small and n_clients is large, it duplicates features randomly to satisfy the minimum.
    """
    random.seed(seed)
    feature_count = len(features)

    # initial base split
    base = feature_count // n_clients
    remainder = feature_count % n_clients
    sizes = [base + 1 if i < remainder else base for i in range(n_clients)]

    # enforcing minimum
    for i in range(n_clients):
        if sizes[i] < min_features:
            sizes[i] = min_features

    # recalculate total required features
    total_needed = sum(sizes)

    # duplicate some features if we need more than we have
    if total_needed > feature_count:
        extended_features = features.copy()
        while len(extended_features) < total_needed:
            extended_features.append(random.choice(features))
    else:
        extended_features = features.copy()

    random.shuffle(extended_features)

    splits = []
    start = 0
    for size in sizes:
        splits.append(extended_features[start:start + size])
        start += size

    return splits
```

```python


def train_vfl_with_clients_min_features(X, Y, features, categorical_columns, clients=4,
                                        epochs=300, batch_size=64, train_test_threshold=0.8):
    """
    Trains the VFL model on a single permutation of features.
    Returns the loss curve and final test accuracy.
    """
    X_train, X_test = X.loc[:int(train_test_threshold * len(X))], X.loc[int(train_test_threshold * len(X)) + 1:]
    Y_train, Y_test = Y.loc[:int(train_test_threshold * len(Y))], Y.loc[int(train_test_threshold * len(Y)) + 1:]

    client_feature_names = split_features_with_minimum(features, clients)

    # map to one-hot encoded features
    encoded_feature_names = list(X.columns)
    updated_client_features = []
    for client_feats in client_feature_names:
        updated = []
        for col in client_feats:
            if col not in categorical_columns:
                updated.append(col)
            else:
                for name in encoded_feature_names:
                    if name.startswith(col + "_"):
                        updated.append(name)
        updated_client_features.append(updated)

    # model architecture
    outs_per_client = 2
    bottom_models = [BottomModel(len(in_feats), outs_per_client * len(in_feats)) for in_feats in
                     updated_client_features]
    Network = VFLNetwork(bottom_models, n_outs=2)

    # capture training logs
    old_stdout = sys.stdout
    sys.stdout = mystdout = io.StringIO()

    # train
    Network.train_with_settings(epochs, batch_size, clients, updated_client_features, X_train, Y_train)

    sys.stdout = old_stdout
    output = mystdout.getvalue()

    # parse loss
    losses = []
    for line in output.split("\n"):
        match = re.search(r"Epoch: \d+ Train accuracy: ([\d.]+)% Loss: ([\d.]+)", line)
        if match:
            losses.append(float(match.group(2)))

    avg_loss = sum(losses) / len(losses)

    # test
    acc, _ = Network.test(X_test, Y_test)

    print(f"Final Train Loss: {losses[-1]:.4f}")
    print(f"Average Train Loss: {avg_loss:.4f}")
    print(f"Test Accuracy: {acc * 100:.2f}%")

    return losses, acc.item()
```

```python
client_cases_expanded = [8, 9, 10] # sanity check
for num_client in (client_cases_expanded):
    splits = split_features_with_minimum(all_features, num_client)
    split_s = []
    for split in splits:
        split_s.append(len(split))

    print(f"split structure for {num_client} clients: {split_s}")
```

```python
losses_all_clients_expanded = []
accuracies_all_clients_expanded = []

client_sizes = [2, 3, 4, 5, 6, 7, 8, 9, 10]
for idx, client_num in enumerate(client_sizes):
    print(f"\nTraining for {client_num} clients...")

    losses, accuracies = train_vfl_with_clients_min_features(
        X=X,
        Y=Y,
        features=all_features,
        categorical_columns=categorical_columns,
        clients=client_num
    )

    losses_all_clients_expanded.append(losses)
    accuracies_all_clients_expanded.append(accuracies)
```

```python
import matplotlib.pyplot as plt

step = 10  # plot every 10th point

plt.figure(figsize=(10, 5))

for i, losses in enumerate(losses_all_clients_expanded):
    downsampled_losses = losses[::step]
    downsampled_epochs = list(range(0, len(losses), step))

    plt.plot(downsampled_epochs, downsampled_losses, label=f'{client_sizes[i]} Clients')

plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.title("Training Loss per Epoch for Different Numbers of Clients with minimum 2 features")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

The minimum feature allocation policy that I enforced guarantees that every client is assigned at least two features, even if that requires duplicating features randomly. While this may increase memory usage and affect communication costs between models, it helps in preserving some correlation between features and reduces the risk of clients learning uninformative or overly sparse representations.

The training results clearly show that enforcing a minimum of two features per client significantly improves convergence and gives a lower final training loss, especially when the number of clients becomes large (e.g. 8, 9, 10). The smoother and faster convergence curves, indicate that this strategy provides stronger local embeddings and better overall collaboration across the federation rather than the first experiment with equal feature partitioning between clients.

This method might not be ideal for strict communication constraints,  butit offers a practical trade-off between performance and resource usage in settings where data dimensionality is limited.

There are better ways for assigning features to clients, for example I checked for using Gini impurity, but that introduces another variable when considering the differences in performance.
